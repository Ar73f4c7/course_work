```python
import pandas as pd
```

```python
df = pd.read_csv('processed_data.csv')
```

```python
df['Группа4']
```

```python
client_sum = df.groupby('Клиент').sum()['Цена']
```

```python
client_sum
```

```python
df = df.set_index('Клиент')
df['ClientSum'] = client_sum
df = df.reset_index()
```

```python
df
```

```python
from tqdm.auto import tqdm
```

```python
pairs = [(row['Клиент'], row['Группа4']) for _, row in tqdm(df.iterrows(), total=len(df))]
```

```python
set_pairs = set(pairs)
```

```python
len(set_pairs)
```

```python
pairs_df = pd.DataFrame(list(set_pairs), columns=['Клиент', 'Товар'])
```

```python
pairs_df
```

```python
un_goods = pd.unique(pairs_df['Товар'])
un_clients = pd.unique(pairs_df['Клиент'])
```

```python
goods_features = df[['Группа2', 'Группа3', 'Группа4', 'Тип', 'Цена']]
```

```python
goods_features.group
```

```python
goods_features = goods_features.groupby('Группа4').first()
```

```python
client_features = df[['Клиент', 'Пол клиента', 'ClientSum']]
```

```python
client_features = client_features.groupby('Клиент').first()
```

```python
client_features
```

```python
import random

negative_pairs = []

for client in un_clients:
    for good in un_goods:
        if (client, good) not in set_pairs:
            if random.random() < 0.01982732450460236:
                negative_pairs.append((client, good))
```

```python
len(negative_pairs)
```

```python
dataset = []

for client, good in set_pairs:
    try:
        features = client_features.loc[client].to_dict()
        for key, value in goods_features.loc[good].iteritems():
            features[key] = value
        features['target'] = 1
        dataset.append(features)
    except:
        pass
    
    
for client, good in negative_pairs:
    try:
        features = client_features.loc[client].to_dict()
        for key, value in goods_features.loc[good].iteritems():
            features[key] = value
        features['target'] = 0
        dataset.append(features)
    except:
        pass
```

```python
dataset = pd.DataFrame(dataset)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(dataset, test_size=0.3)
```

```python
import catboost as cb
```

```python
train
```

```python
model = cb.CatBoostClassifier(cat_features=['Пол клиента', 'Группа2', 'Группа3', 'Тип'], eval_metric='BalancedAccuracy')
```

```python
X_train = train.drop(['target'], axis=1)
X_test = test.drop(['target'], axis=1)

y_train = train['target']
y_test = test['target']
```

```python
model.fit(X_train, y_train, eval_set=(X_test, y_test))
```

```python
len(un_clients) * len(un_goods)
```

```python
pred_ds = []
pred_pairs = []
for client in tqdm(un_clients):
    for good in un_goods:
        if (client, good) not in set_pairs:
            try:
                features = client_features.loc[client].to_dict()
                for key, value in goods_features.loc[good].iteritems():
                    features[key] = value
                features['target'] = 0
                pred_ds.append(features)
                pred_pairs.append((client, good))
            except:
                continue
```

```python
pred_ds_df = pd.DataFrame(pred_ds).drop(['target'], axis=1)
```

```python
pred_ds_df
```

```python
probas = model.predict_proba(pred_ds_df)
```

```python
pred_ds_df['Вероятность отклика'] = probas[:, 1]
```

```python
pred_ds_df['Клиент'] = [client for client, good in pred_pairs]
```

```python
pred_ds_df['Группа4'] = [good for client, good in pred_pairs]
```

```python
pred_ds_df.sort_values(['Вероятность отклика'], ascending=False).head(100).to_excel('content_based.xlsx', index=None)
```
